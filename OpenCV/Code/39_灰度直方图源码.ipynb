```python
"""
    灰度直方图 源码
    直方图：本质， 统计每个像素灰度出现的概率， 0-255, p概率
"""

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("..\Images\m5m.jpg",1)
imgInfo = img.shape
height = imgInfo[0]
width = imgInfo[1]

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
count = np.zeros(256,np.float)

for i in range(0,height):
    for j in range(0,width):
        pixel = gray[i,j]
        index = int(pixel)
        count[index] += 1

for i in range(0,255):
    count[i] = count[i]/(height*width)

x = np.linspace(0,255,256)
y = count
plt.bar(x,y,0.9,alpha = 1,color = "b")
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
"""
    灰度直方图 均衡化
    直方图：本质， 统计每个像素灰度出现的概率， 0-255, p概率
    
    累计概率：
    1  - >  0.2    0.2
    2  - >  0.3    0.5
    3  - >  0.1    0.6
    ...  ...   ...
    100 - > p  ==>  可通过 255*p = new 作为 100的 新的映射， 即 均衡化
"""

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("..\Images\m5m.jpg",1)
imgInfo = img.shape
height = imgInfo[0]
width = imgInfo[1]

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
count = np.zeros(256,np.float)

for i in range(0,height):
    for j in range(0,width):
        pixel = gray[i,j]
        index = int(pixel)
        count[index] += 1

for i in range(0,255):
    count[i] = count[i]/(height*width)

# 计算累计概率
sum1 = float(0)
for i in range(0,256):
    sum1 += count[i]
    count[i] = sum1
# print(count)
# 计算映射表
map1 = np.zeros(256,np.uint16)
for i range(0,256):
    map1[i] = np.uint16(count[i]*255)

#映射
for i in range(0,height):
    for j in range(0,width):
        pixel = gray[i,j]
        gray[i,j] = map1[pixel]
cv2.imshow("dst",gray)
cv2.waitKey(0)
cv2.destroyWindows()
```
