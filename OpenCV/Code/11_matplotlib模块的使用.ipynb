```python
"""折线图"""
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5,6,4,1,3,6,10,14])

plt.plot(x,y,'r') # 折线  参数1：x轴， 参数2：y轴， 参数3：color颜色
```

```python
"""折线图"""
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5,6,4,1,3,6,10,14])

plt.plot(x,y,"g",lw=3) #参数1：x轴， 参数2：y轴， 参数3：color颜色， 参数4：线条宽度
```

```python
"""柱状图"""
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5,6,4,1,3,6,10,14])

plt.bar(x,y,0.5,alpha=1,color="b") #参数1: x轴， 参数2：y轴， 参数3：每个柱状图占用的宽度， 参数4：透明度， 参数5：颜色
plt.show()
```
