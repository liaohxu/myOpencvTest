```python
# 1，灰度图像 最重要 2，图片处理的基础  3，实时性
# 定点 快于 浮点类型    + - 快于 * / 方法
# r*0.299 + g*0.587 + b*0.114

# 比如：灰度处理 方法三
import cv2
import numpy as np

# 读取彩色图片
img = cv2.imread("..\Images\m3m.jpg",1)
imgInfo = img.shape
height = imgInfo[0]
width = imgInfo[1]
deep = imgInfo[2]

#创建一个 相同尺寸的 全空矩阵
dst = np.zeros((height,width,deep),np.uint8)

# 通过循环 填充 每一个像素点
for i in range(0,height):
    for j in range(0,width):
        # 获取每个像素点的 b,g,r
        (b,g,r) = img[i,j]
        # 以均值的形式 求 gray值
        gray = (int(b)+int(g)+int(r))/3
        # 将 gray值 赋值给 新图像的 每个像素点
        dst[i,j] = np.uint8(gray)
    
# 显示与关闭窗口
cv2.imshow("dst",dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
        
```

```python
# 优化方法，将 均值算法公式 转为 定点【整形】,会有一点误差，但是不大
# 在这里 放大 4倍， 也可以方法 100倍，1000倍，越大，精度越高

import cv2
import numpy as np

# 读取彩色图片
img = cv2.imread("..\Images\m3m.jpg",1)
imgInfo = img.shape
height = imgInfo[0]
width = imgInfo[1]
deep = imgInfo[2]

#创建一个 相同尺寸的 全空矩阵
dst = np.zeros((height,width,deep),np.uint8)

# 通过循环 填充 每一个像素点
for i in range(0,height):
    for j in range(0,width):
        # 获取每个像素点的 b,g,r
        (b,g,r) = img[i,j]
        b = int(b)
        g = int(g)
        r = int(r)
        
        #gray = (b*1 + g*2 + r*1)/4  # 通过 扩大倍数 在 缩小到 原倍数
        
        gray = (b*20 + g*40 + r*20)/80  # 通过 扩大倍数 在 缩小到 原倍数, 倍数放大越高， 精度越准
        
        #gray = (b + (g<<1) + r)>>2  # (g<<1) 表示 左移动 1位， ()>>2 表示 右移动 2位
        
        # 将 gray值 赋值给 新图像的 每个像素点
        dst[i,j] = np.uint8(gray)
    
# 显示与关闭窗口
cv2.imshow("dst",dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
